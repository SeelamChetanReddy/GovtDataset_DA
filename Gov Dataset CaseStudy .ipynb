1.Total Schools by State/UT
This bar chart shows the total number of schools in each State/UT.
It helps compare the distribution of schools across different regions.

16. Correlation Matrix of School Data
This heatmap visualizes the correlation between various numerical features in the dataset.
It helps identify strong relationships between different school-related factors, such as total schools and computer facility availability.

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="Blues", linewidths=0.5)
plt.title("Correlation Matrix of School Data")
plt.show()
```

17. Pair Plot
This pair plot displays scatter plots and KDE distributions for selected numerical variables related to government and private schools.
It helps identify relationships, trends, and distributions among school counts and computer facility percentages.

```python
sns.pairplot(df, vars=["Total Schools - Govt.", "Total Schools - Pvt. Unaided", "% of Schools with Computer Facility - Govt.", "% of Schools with Computer Facility - Pvt. Unaided"], diag_kind='kde')
plt.show()
```

18. KDE Plot
This KDE plot visualizes the distribution of schools with computer facilities across all management types.
It helps understand the density and spread of schools equipped with computer infrastructure.

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df["Schools with Computer Facility - All Management"], fill=True, color="blue")
plt.title("KDE Plot of Schools with Computer Facility")
plt.xlabel("Number of Schools")
plt.show()
```

19. Violin Plot of Total Schools - All Management
This violin plot visualizes the distribution of total schools across different States/UTs.
It combines a box plot and a density plot to show both the spread and concentration of school counts.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(y=df["Total Schools - All Management"], color="skyblue")
plt.title("Violin Plot for Total Schools Distribution")
plt.show()
```

20. Line Plot
This line plot shows the trend of schools with computer facilities across different States/UTs.
It helps visualize variations in digital infrastructure availability among schools.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df.index, y=df["Schools with Computer Facility - All Management"], marker="o", color="blue")
plt.title("Trend of Schools with Computer Facility")
plt.xlabel("Index")
plt.ylabel("Schools with Computer Facility")
plt.show()
```

# Analysis Description

**Government School Infrastructure Analysis Report**

**1. Introduction**
This report presents an analysis of school infrastructure across various States and Union Territories (UTs) in India, using data from the UDISE 2020-21 dataset. The focus is on school distribution, digital infrastructure, and government versus private school comparisons.

**2. Key Findings**

- **Total Schools by State/UT:** The highest number of schools is observed in Uttar Pradesh, Maharashtra, and West Bengal, while smaller states and UTs have significantly fewer schools.
- **Government vs. Private Schools:** Government schools dominate in rural areas, whereas private unaided schools are more prevalent in urban centers.
- **Computer Facility Distribution:** A disparity exists in access to computer facilities, with private schools having better digital infrastructure compared to government schools in many states.
- **State-Wise Digital Readiness:** States like Delhi and Kerala have a higher percentage of schools with computer facilities, while states such as Bihar and Jharkhand lag behind.
- **Correlation Analysis:** A strong correlation is observed between total school count and government school count, while private schools show varied distribution patterns.
